```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data
data = {
    "Shopot": {
        "1_fr_noisy": {"wer": 0.3829, "time": float("inf")},
        "2_fr_slow": {"wer": 0.3648, "time": float("inf")},
        "3_en_windy": {"wer": 0.3928, "time": float("inf")},
        "4_en_noisy": {"wer": 0.2430, "time": float("inf")},
        "5_es_noisy": {"wer": 0.3576, "time": float("inf")},
        "6_ru_noisy": {"wer": 0.4739, "time": float("inf")},
        "7_de_noisy": {"wer": 0.2613, "time": float("inf")},
        "8_ru_noisy": {"wer": 1.0000, "time": float("inf")},
        "9_ru_noisy": {"wer": 0.4700, "time": float("inf")},
        "10_ru_song": {"wer": 1.0000, "time": float("inf")},
        "11_en_song": {"wer": 0.9808, "time": float("inf")},
        "12_fr_song": {"wer": 0.9897, "time": float("inf")},
        "13_ru_interview": {"wer": 0.2938, "time": float("inf")},
        "14_en_kurz": {"wer": 1.0000, "time": float("inf")},
        "15_en_ltt": {"wer": 0.0714, "time": float("inf")},
        "16_en_mustard": {"wer": 0.1171, "time": float("inf")},
    },
    "AssemblyAI": {
        "1_fr_noisy": {"wer": 0.9969, "time": 8.05},
        "2_fr_slow": {"wer": 0.9669, "time": 9.51},
        "3_en_windy": {"wer": 0.3358, "time": 8.97},
        "4_en_noisy": {"wer": 0.2126, "time": 8.78},
        "5_es_noisy": {"wer": 0.9883, "time": 10.05},
        "6_ru_noisy": {"wer": 1.0000, "time": 4.41},
        "7_de_noisy": {"wer": 0.9520, "time": 7.93},
        "8_ru_noisy": {"wer": 0.9993, "time": 13.72},
        "9_ru_noisy": {"wer": 1.0000, "time": 4.62},
        "10_ru_song": {"wer": 1.0000, "time": 10.61},
        "11_en_song": {"wer": 0.7436, "time": 8.21},
        "12_fr_song": {"wer": 1.0000, "time": 10.25},
        "13_ru_interview": {"wer": 1.0000, "time": 21.32},
        "14_en_kurz": {"wer": 0.1862, "time": 37.461},
        "15_en_ltt": {"wer": 0.1215, "time": 21.16},
        "16_en_mustard": {"wer": 0.1385, "time": 19.34},
    },
    "OpenAI Whisper API": {
        "1_fr_noisy": {"wer": 0.4045, "time": 34.59},
        "2_fr_slow": {"wer": 0.3862, "time": 35.03},
        "3_en_windy": {"wer": 0.2986, "time": 15.40},
        "4_en_noisy": {"wer": 0.2080, "time": 25.95},
        "5_es_noisy": {"wer": 0.4045, "time": 20.77},
        "6_ru_noisy": {"wer": 0.4572, "time": 11.85},
        "7_de_noisy": {"wer": 0.2669, "time": 16.13},
        "8_ru_noisy": {"wer": 0.2035, "time": 42.25},
        "9_ru_noisy": {"wer": 0.4700, "time": 29.99},
        "10_ru_song": {"wer": 0.9392, "time": 4.45},
        "11_en_song": {"wer": 1.0000, "time": 5.17},
        "12_fr_song": {"wer": 0.5724, "time": 13.34},
        "13_ru_interview": {"wer": 1, "time": float("inf")},
        "14_en_kurz": {"wer": 1, "time": float("inf")},
        "15_en_ltt": {"wer": 1, "time": float("inf")},
        "16_en_mustard": {"wer": 0.1082, "time": 28.22},
    },
    "Fireworks Large-v3": {
        "1_fr_noisy": {"wer": 0.3760, "time": 5.49},
        "2_fr_slow": {"wer": 0.3676, "time": 6.47},
        "3_en_windy": {"wer": 0.3073, "time": 3.98},
        "4_en_noisy": {"wer": 0.2080, "time": 5.25},
        "5_es_noisy": {"wer": 0.3107, "time": 5.71},
        "6_ru_noisy": {"wer": 0.4551, "time": 3.87},
        "7_de_noisy": {"wer": 0.2006, "time": 4.62},
        "8_ru_noisy": {"wer": 0.1754, "time": 7.73},
        "9_ru_noisy": {"wer": 0.4463, "time": 10.39},
        "10_ru_song": {"wer": 0.9227, "time": 3.56},
        "11_en_song": {"wer": 1.0000, "time": 2.93},
        "12_fr_song": {"wer": 1.0000, "time": 3.32},
        "13_ru_interview": {"wer": 0.2361, "time": 13.04},
        "14_en_kurz": {"wer": 0.2869, "time": 24.20},
        "15_en_ltt": {"wer": 0.0730, "time": 13.08},
        "16_en_mustard": {"wer": 0.1265, "time": 7.76},
    },
    "Fireworks Large-v3 Turbo": {
        "1_fr_noisy": {"wer": 0.3821, "time": 3.449},
        "2_fr_slow": {"wer": 0.3807, "time": 4.287},
        "3_en_windy": {"wer": 0.3098, "time": 3.548},
        "4_en_noisy": {"wer": 0.2096, "time": 3.618},
        "5_es_noisy": {"wer": 0.3494, "time": 3.389},
        "6_ru_noisy": {"wer": 0.5825, "time": 2.996},
        "7_de_noisy": {"wer": 0.2105, "time": 2.422},
        "8_ru_noisy": {"wer": 0.1839, "time": 3.943},
        "9_ru_noisy": {"wer": 0.4452, "time": 3.212},
        "10_ru_song": {"wer": 0.9558, "time": 3.487},
        "11_en_song": {"wer": 1.0000, "time": 2.812},
        "12_fr_song": {"wer": 1.0000, "time": 2.957},
        "13_ru_interview": {"wer": 0.2528, "time": 5.022},
        "14_en_kurz": {"wer": 0.1925, "time": 7.662},
        "15_en_ltt": {"wer": 0.0742, "time": 16.73},
        "16_en_mustard": {"wer": 0.1296, "time": 4.665},
    },
    "Nexara": {
        "1_fr_noisy": {"wer": 0.4607, "time": 4.12},
        "2_fr_slow": {"wer": 0.3317, "time": 4.55},
        "3_en_windy": {"wer": 0.3222, "time": 3.28},
        "4_en_noisy": {"wer": 0.2156, "time": 4.08},
        "5_es_noisy": {"wer": 0.3505, "time": 4.29},
        "6_ru_noisy": {"wer": 0.4656, "time": 3.06},
        "7_de_noisy": {"wer": 0.2331, "time": 3.92},
        "8_ru_noisy": {"wer": 0.3796, "time": 7.14},
        "9_ru_noisy": {"wer": 0.4452, "time": 4.14},
        "10_ru_song": {"wer": 0.7403, "time": 3.03},
        "11_en_song": {"wer": 0.7500, "time": 2.36},
        "12_fr_song": {"wer": 0.6862, "time": 3.57},
        "13_ru_interview": {"wer": 0.2673, "time": 9.53},
        "14_en_kurz": {"wer": 0.1969, "time": 15.77},
        "15_en_ltt": {"wer": 0.0888, "time": 10.01},
        "16_en_mustard": {"wer": 0.1474, "time": 25.83},
    },
    "DeepInfra Whisper v3 Large": {
        "1_fr_noisy": {"wer": 0.4076, "time": 9.21},
        "2_fr_slow": {"wer": 1.0000, "time": float("inf")},
        "3_en_windy": {"wer": 0.3271, "time": 5.76},
        "4_en_noisy": {"wer": 0.2521, "time": 8.94},
        "5_es_noisy": {"wer": 0.3857, "time": 7.84},
        "6_ru_noisy": {"wer": 0.4969, "time": 4.95},
        "7_de_noisy": {"wer": 0.2712, "time": 7.35},
        "8_ru_noisy": {"wer": 1.0000, "time": float("inf")},
        "9_ru_noisy": {"wer": 1.0000, "time": float("inf")},
        "10_ru_song": {"wer": 1.0000, "time": float("inf")},
        "11_en_song": {"wer": 0.8397, "time": 4.68},
        "12_fr_song": {"wer": 1.0000, "time": float("inf")},
        "13_ru_interview": {"wer": 0.2881, "time": 21.75},
        "14_en_kurz": {"wer": 1.0000, "time": float("inf")},
        "15_en_ltt": {"wer": 0.1286, "time": 18.49},
        "16_en_mustard": {"wer": 0.1166, "time": 11.13},
    },
}

file_durations = {
    "1_fr_noisy": 544.176,
    "2_fr_slow": 744.816,
    "3_en_windy": 280.296,
    "4_en_noisy": 556.32,
    "5_es_noisy": 449.232,
    "6_ru_noisy": 164.76,
    "7_de_noisy": 316.152,
    "8_ru_noisy": 765.192,
    "9_ru_noisy": 466.68,
    "10_ru_song": 237.627,
    "11_en_song": 224.387,
    "12_fr_song": 328.987,
    "13_ru_interview": 1687.44,
    "14_en_kurz": 3810.65,
    "15_en_ltt": 1432.872,
    "16_en_mustard": 861.042,
}

def create_mobile_friendly_charts(show_values=True):
    # Get providers and audio files
    providers = list(data.keys())
    audio_files = list(data[providers[0]].keys())
    
    # Split audio files into two groups: files 1-8 and files 9-16
    first_half = audio_files[:8]
    second_half = audio_files[8:16]
    
    # Define a color map for providers
    colors = plt.cm.tab10(np.linspace(0, 1, len(providers)))
    
    # Create two plots
    plots = []
    
    # Plot 1: WER for both halves (files 1-8 and 9-16)
    fig1, (ax1_top, ax1_bottom) = plt.subplots(2, 1, figsize=(12, 10))
    
    # Plot 2: Speed for both halves (files 1-8 and 9-16)
    fig2, (ax2_top, ax2_bottom) = plt.subplots(2, 1, figsize=(12, 10))
    
    # Set width of bars
    bar_width = 0.8 / len(providers)
    
    # Function to plot WER data for a specific set of files to a specific axis
    def plot_wer(ax, file_group, group_label):
        indices = np.arange(len(file_group))
        
        for i, provider in enumerate(providers):
            wer_values = []
            for audio in file_group:
                wer = data[provider][audio]["wer"]
                wer_values.append(wer if wer != 1.0 else 0.98)  # Cap at 0.98 for visibility
            
            positions = indices + bar_width * i - bar_width * (len(providers) - 1) / 2
            bars = ax.bar(positions, wer_values, bar_width, label=provider, color=colors[i])
            
            if show_values:
                for j, bar in enumerate(bars):
                    height = bar.get_height()
                    value_text = f"{data[provider][file_group[j]]['wer']:.2f}"
                    ax.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                           value_text, ha='center', va='bottom', rotation=90, fontsize=8)
        
        # Set titles and labels
        ax.set_title(f'Word Error Rate (WER) - {group_label} ( - )', fontsize=14)
        ax.set_ylabel('WER', fontsize=12)
        ax.set_ylim(0, 1.05)  # Set y-axis limit for WER
        
        # Format x-axis labels
        formatted_labels = [f"{audio}\n{file_durations[audio]:.0f}s" for audio in file_group]
        ax.set_xticks(indices)
        ax.set_xticklabels(formatted_labels, rotation=45, ha='right', fontsize=10)
        ax.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Function to plot Speed data for a specific set of files to a specific axis
    def plot_speed(ax, file_group, group_label):
        indices = np.arange(len(file_group))
        
        for i, provider in enumerate(providers):
            speed_values = []
            for audio in file_group:
                time = data[provider][audio]["time"]
                duration = file_durations[audio]
                
                if time == float("inf") or time <= 0:
                    speed = 0  # Can't process or no data
                else:
                    speed = duration / time
                
                speed_values.append(speed)
            
            positions = indices + bar_width * i - bar_width * (len(providers) - 1) / 2
            bars = ax.bar(positions, speed_values, bar_width, label=provider, color=colors[i])
            
            if show_values:
                for j, bar in enumerate(bars):
                    height = bar.get_height()
                    if height > 0:
                        value_text = f"{speed_values[j]:.2f}x"
                        ax.text(bar.get_x() + bar.get_width()/2., height + 0.5,
                               value_text, ha='center', va='bottom', rotation=90, fontsize=8)
        
        # Set titles and labels
        ax.set_title(f'  - {group_label} ( - )', fontsize=14)
        ax.set_ylabel(' ( / )', fontsize=12)
        
        # Format x-axis labels
        formatted_labels = [f"{audio}\n{file_durations[audio]:.0f}s" for audio in file_group]
        ax.set_xticks(indices)
        ax.set_xticklabels(formatted_labels, rotation=45, ha='right', fontsize=10)
        ax.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Plot WER data
    plot_wer(ax1_top, first_half, " 1-8")
    plot_wer(ax1_bottom, second_half, " 9-16")
    
    # Add legend to figure 1 (only once)
    handles, labels = ax1_top.get_legend_handles_labels()
    fig1.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, 0.98), 
                ncol=4, fancybox=True, shadow=True, fontsize=10)
    
    # Plot Speed data
    plot_speed(ax2_top, first_half, " 1-8")
    plot_speed(ax2_bottom, second_half, " 9-16")
    
    # Add legend to figure 2 (only once)
    handles, labels = ax2_top.get_legend_handles_labels()
    fig2.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, 0.98), 
                ncol=4, fancybox=True, shadow=True, fontsize=10)
    
    # Adjust layout
    for fig in [fig1, fig2]:
        plt.figure(fig.number)
        plt.tight_layout()
        plt.subplots_adjust(top=0.9, hspace=0.4)
    
    # Save figures
    fig1.savefig('wer_comparison.png', dpi=300, bbox_inches='tight')
    fig2.savefig('speed_comparison.png', dpi=300, bbox_inches='tight')
    
    return [(fig1, 'wer_comparison.png'), (fig2, 'speed_comparison.png')]

# Call with value display setting
show_values = True  # Set to False to disable value labels
plots = create_mobile_friendly_charts(show_values)

# Display all plots
for fig, _ in plots:
    plt.figure(fig.number)
    plt.show()
```
